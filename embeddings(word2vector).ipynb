```python
!pip install pypdf gensim nltk

from pypdf import PdfReader
from gensim.models import Word2Vec
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

# Read PDF
reader = PdfReader("/content/resumes/data_science_02_sarah_williams.pdf")

text = ""
for page in reader.pages:
    if page.extract_text():
        text += page.extract_text()

# Convert text into sentences
sentences = nltk.sent_tokenize(text)

# Convert each sentence into words
tokenized_sentences = [
    nltk.word_tokenize(sentence.lower())
    for sentence in sentences
]

print(tokenized_sentences[:3])
model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,
    window=5,
    min_count=1
)
model.save("pdf_word2vec.model")
print(model.wv["data"])
model.wv.most_similar("data")
```

```python
!pip install pypdf gensim nltk

import os
import glob
from pypdf import PdfReader
from gensim.models import Word2Vec
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

# Folder containing your PDFs
import os, glob

folder_path = "/content/resumes"
print("Contents of folder:")
for f in os.listdir(folder_path):
    print("  ", f)

pdf_files = glob.glob(os.path.join(folder_path, "*.pdf"))
print("\nMatched PDFs:", pdf_files)

for pdf_path in pdf_files:
    print(f"Reading: {os.path.basename(pdf_path)}")
    try:
        reader = PdfReader(pdf_path)
    except Exception as e:
        print(f"  Skipped (could not open): {e}")
        continue

    text = ""
    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text

    # Split into sentences, then into words
    for sentence in nltk.sent_tokenize(text):
        tokenized_sentences.append(nltk.word_tokenize(sentence.lower()))

print(f"\nTotal sentences collected: {len(tokenized_sentences)}")
print("Sample:", tokenized_sentences[:3])

# Train one model on everything
model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=50,
    window=5,
    min_count=1
)

model.save("pdf_word2vec.model")
print(model.wv["engineering"])
print(model.wv.most_similar("engineering"))
```
